1.	Импортировал библиотеку pandas как pd. Загрузил датасет bookings.csv с разделителем ;. Проверил размер таблицы, типы переменных, а затем вывел первые 7 строк, чтобы посмотреть на данные.

```python
import pandas as pd
bookings = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/2_bookings.csv',
                       encoding = 'windows-1251',
                       sep = ';')
rows, cols = bookings.shape
# print(f'rows={rows},cols={cols}')
bookings.dtypes
bookings_head = bookings.head(7)
bookings.columns = bookings.columns.str.lower().str.replace(' ','_')
```

2.	Привел названия колонок к нижнему регистру и заменил пробелы на знак нижнего подчеркивания.

```python
# Можно сделать переименование и добавление нижнего регистра

rename_cols = {}
for col in bookings.columns:
  rename_cols[col] = col.replace(' ','_').lower()
```

3.	Пользователи из каких стран совершили наибольшее число успешных бронирований? Укажите топ-5.

```python
result = bookings[bookings['is_canceled']==0].\
groupby('country', as_index=False).\
aggregate({'is_canceled':'count'}).\
sort_values('is_canceled',ascending = False).head(5)
print(result)
```

9.	Создал переменную has_kids, которая принимает значение True, если клиент при бронировании указал хотя бы одного ребенка (total_kids), в противном случае – False. Посчитал отношение количества ушедших пользователей к общему количеству клиентов, выраженное в процентах (churn rate). Указал, среди какой группы показатель выше.

```python
def has_kids(x):
  if x > 0:
    return True
  else:
    return False
```

```python
bookings['is_canceled'].value_counts()
```

```python
bookings['has_kids'] = bookings.total_kids.apply(has_kids)
```

```python
#Можно сделать вот так
bookings['has_kids'] = bookings.total_kids.astype(bool)
print(bookings['has_kids'])
```

```python
bookings.has_kids.value_counts()
```

```python
bookings[['has_kids','is_canceled']].value_counts()
```

```python
bookings[['has_kids','is_canceled']].value_counts()/bookings['has_kids'].value_counts()*100
```
